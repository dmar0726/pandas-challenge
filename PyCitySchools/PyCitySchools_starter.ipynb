```python
# Dependencies
import pandas as pd

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student files
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total schools
total_schools = school_data["School ID"].count()
# Total students
total_students = student_data["Student ID"].count()
# Total budget
total_budget = school_data["budget"].sum()
# Average math Score
average_math = student_data["math_score"].mean()/100
# Average reading Score
average_reading = student_data["reading_score"].mean()/100
# Percent of students passing math
passing_math = student_data.loc[student_data["math_score"]>=70]["math_score"].count()/total_students
# Percent of students passing reading
passing_reading = student_data.loc[student_data["reading_score"]>=70]["reading_score"].count()/total_students
# Overall passing rate
overall_passing = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)]['student_name'].count()/total_students
district_summary = pd.DataFrame({"Total Schools":[total_schools],
                        "Total Students":[total_students],
                        "Total Budget":[total_budget],
                        "Average Math Score":[average_math],
                        "Average Reading Score":[average_reading],
                        "% Passing Math":[passing_math],
                        "% Passing Reading":[passing_reading],
                        "Overall Passing Rate":[overall_passing]                            
                                })
# Formatting
district_summary['Total Students'] = district_summary['Total Students'].map("{:,}".format)
district_summary['Total Budget'] = district_summary['Total Budget'].map("${:,.2f}".format)
district_summary['Average Math Score'] = district_summary['Average Math Score'].map("{:.1%}".format)
district_summary['Average Reading Score'] = district_summary['Average Reading Score'].map("{:.1%}".format)
district_summary['% Passing Math'] = district_summary['% Passing Math'].map("{:.1%}".format)
district_summary['% Passing Reading'] = district_summary['% Passing Reading'].map("{:.1%}".format)
district_summary['Overall Passing Rate'] = district_summary['Overall Passing Rate'].map("{:.1%}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Grouped by school name 
grouped_school = school_data_complete.groupby(['school_name'])
# Total students for each school
total_student = grouped_school.size()
# Get the school type for each school
school_type = grouped_school['type'].first()
# Total budget for each school
total_budget = grouped_school['budget'].first()
# Budget per student for each school
budget_per_student = total_budget/total_student
# Average math score for each school
average_math_score = grouped_school['math_score'].mean()/100
# Average reading score for each school
average_reading_score = grouped_school['reading_score'].mean()/100
# Percentange of passing math for each school
grouped_passing_math = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
percent_passing_math = (grouped_passing_math/total_student)
# Percentange of passing reading for each school
grouped_passing_reading = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
percent_passing_reading = (grouped_passing_reading/total_student)
# Calculate the overall passing for each school
overall_passing = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]
grouped_overall_passing = overall_passing.groupby(['school_name']).count()['student_name'] / total_student

#Create a dataframe to hold the above results
school = {
    'School Type': school_type,
    'Total Students':total_student,
    'Total School Budget': total_budget,
    'Per Student Budget': budget_per_student,
    'Average Math Score': average_math_score,
    'Average Reading Score': average_reading_score,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    '% Overall Passing Rate': grouped_overall_passing,
}
school_summary = pd.DataFrame(school)

# Copy of dataframe
displayed_school_summary = school_summary.copy()

# Formatting
displayed_school_summary['Per Student Budget'] = displayed_school_summary['Per Student Budget'].map('${:,.2f}'.format)
displayed_school_summary['Total School Budget'] = displayed_school_summary['Total School Budget'].map('${:,.2f}'.format)
displayed_school_summary['Total Students'] = displayed_school_summary['Total Students'].map("{:,}".format)
displayed_school_summary['Average Math Score'] = displayed_school_summary['Average Math Score'].map("{:.1%}".format)
displayed_school_summary['Average Reading Score'] = displayed_school_summary['Average Reading Score'].map("{:.1%}".format)
displayed_school_summary['% Passing Math'] = displayed_school_summary['% Passing Math'].map("{:.1%}".format)
displayed_school_summary['% Passing Reading'] = displayed_school_summary['% Passing Reading'].map("{:.1%}".format)
displayed_school_summary['% Overall Passing Rate'] = displayed_school_summary['% Overall Passing Rate'].map("{:.1%}".format)

displayed_school_summary
```

```python
# Sort and display the five best schools
top_performing_schools = displayed_school_summary.sort_values(by='% Overall Passing Rate',ascending=False)
top_performing_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# 9th grade at each school
school_avg_math_9th = school_data_complete[school_data_complete['grade']=='9th'].groupby('school_name')['math_score'].mean()/100
# 10th grade at each school
school_avg_math_10th = school_data_complete[school_data_complete['grade']=='10th'].groupby('school_name')['math_score'].mean()/100
# 11th grade at each school
school_avg_math_11th = school_data_complete[school_data_complete['grade']=='11th'].groupby('school_name')['math_score'].mean()/100
# 12th grade at each school
school_avg_math_12th = school_data_complete[school_data_complete['grade']=='12th'].groupby('school_name')['math_score'].mean()/100

# Create a dataframe to hold the above results
grade_math_score={
    '9th':school_avg_math_9th,
    '10th':school_avg_math_10th,
    '11th':school_avg_math_11th,
    '12th':school_avg_math_12th,
    }

math_score_by_grade = pd.DataFrame(grade_math_score)
# Formatting
math_score_by_grade['9th'] = math_score_by_grade['9th'].map("{:.1%}".format)
math_score_by_grade['10th'] = math_score_by_grade['10th'].map("{:.1%}".format)
math_score_by_grade['11th'] = math_score_by_grade['11th'].map("{:.1%}".format)
math_score_by_grade['12th'] = math_score_by_grade['12th'].map("{:.1%}".format)

math_score_by_grade.head(15)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Bins
spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$585", "$585-630", "$630-645", "$645-675"]

# Create a new data frame by locating the desired columns
scores_spending = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','% Overall Passing Rate',]]
# Add a new column
scores_spending['Spending Ranges (Per Student)']= pd.cut(school_summary['Per Student Budget'],spending_bins,labels=group_names)
# Group by spending
scores_spending = scores_spending.groupby('Spending Ranges (Per Student)').mean()
# Formatting
scores_spending['Average Math Score'] = scores_spending['Average Math Score'].map("{:.1%}".format)
scores_spending['Average Reading Score'] = scores_spending['Average Reading Score'].map("{:.1%}".format)
scores_spending['% Passing Math'] = scores_spending['% Passing Math'].map("{:.1%}".format)
scores_spending['% Passing Reading'] = scores_spending['% Passing Reading'].map("{:.1%}".format)
scores_spending['% Overall Passing Rate'] = scores_spending['% Overall Passing Rate'].map("{:.1%}".format)

scores_spending.head()
```
